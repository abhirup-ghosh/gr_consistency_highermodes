```python
import pycbc
import pycbc.filter.matchedfilter as mfilter
import pycbc.psd
import pycbc.noise.gaussian
from pycbc  import  detector
import numpy as np
import matplotlib.pyplot as plt
import lal
```

```python
# signal parameters
f_low=20.

m1, m2 = 8.4, 1.4
M=9.8          # in MSUN
q=1./6         
Mc=(M*q**0.6)/((1.+q)**1.2)
SNR_req=25.    
iota=0.
Psi_ref=np.pi
t0=6.

ra=1.          
dec =1.
pol=-np.pi/2.
```

```python
t_geom, hp_geom, hc_geom = np.loadtxt('../data/polarizations/NRPolzns_BBH_SpEC_q6.00_spin1[0.00,0.00,-0.00]_spin2[-0.00,-0.00,-0.00]_iota_0.00_psi_0.00.dat.gz', unpack=True)

dist = 1. #in Mpc
t_SI = t_geom * (M*lal.MTSUN_SI)
dt_SI = t_SI[1]-t_SI[0]
hp_SI = hp_geom*(M*lal.MTSUN_SI)/(dist*1e6*lal.PC_SI/lal.C_SI)
hc_SI = hc_geom*(M*lal.MTSUN_SI)/(dist*1e6*lal.PC_SI/lal.C_SI)

hpf_SI = np.fft.fft(hp_SI)*dt_SI
hcf_SI = np.fft.fft(hc_SI)*dt_SI
f_SI = np.fft.fftfreq(len(hpf_SI), d=dt_SI)
df = f_SI[1] - f_SI[0]

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(t_SI, hp_SI, 'r')
plt.plot(t_SI, hc_SI, 'k--')

plt.subplot(122)
plt.loglog(f_SI, abs(hpf_SI), 'r')
plt.loglog(f_SI, abs(hcf_SI), 'k--')

plt.show()
```

```python
Fp,Fc = detector.overhead_antenna_pattern(ra, dec, pol)
psd = pycbc.psd.aLIGOZeroDetHighPower(len(hpf_SI), df, f_low)

signal=Fp*hpf_SI+Fc*hcf_SI
signal_freq=pycbc.types.frequencyseries.FrequencySeries(signal,delta_f=df,dtype=complex)
SNR=mfilter.sigma(signal_freq,psd=psd,low_frequency_cutoff=f_low)

print 'initial SNR:%f'%SNR

r = SNR/SNR_req

hp_SI = hp_geom*(M*lal.MTSUN_SI)/(r*1e6*lal.PC_SI/lal.C_SI)
hc_SI = hc_geom*(M*lal.MTSUN_SI)/(r*1e6*lal.PC_SI/lal.C_SI)

hpf_SI = np.fft.fft(hp_SI)*dt_SI
hcf_SI = np.fft.fft(hc_SI)*dt_SI

signal=Fp*hpf_SI+Fc*hcf_SI
signal_freq=pycbc.types.frequencyseries.FrequencySeries(signal,delta_f=df,dtype=complex)
SNR=mfilter.sigma(signal_freq,psd=psd,low_frequency_cutoff=f_low)

print 'rescaled distance: %f Mpc for a fixed SNR: %f'%(r, SNR)

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.plot(t_SI, hp_SI, 'r')
plt.plot(t_SI, hc_SI, 'k--')

plt.subplot(122)
plt.loglog(f_SI, abs(hpf_SI), 'r')
plt.loglog(f_SI, abs(hcf_SI), 'k--')
plt.loglog(f_SI, abs(signal), 'b--')
plt.loglog(f_SI, psd**0.5, 'g--')

plt.show()
```

```python
##### Output location and data file name #######

'''
Description of output data:

Output data contains [freq ,Real.part of signal ,Imag.part of signal ,PSD] in the same order
Data is non-zero starting from f_low. Last entry of data is (N-2)*df

'''
out_file = '../injections/modGR_simulations/NSBH_SXS/GR_M_%.2f_Mc_%.2f_q_%.2f_snr_25_i_%.2f_dL_%.2f_phiref_%.2f_t0_%.2f_ra_%.2f_dec_%.2f_psi_%.2f_flow_20Hz_NSBH_SXS.dat'%(M,Mc,q,iota, r, Psi_ref, t0, ra, dec, pol)

### Generating noise from psd ###
#noise=pycbc.noise.gaussian.frequency_noise_from_psd(psd, seed=None)
data=signal#+noise ## comment noise to generate noise free data
#################################

datar=np.real(data)
datai=np.imag(data)

plt.figure(figsize=(8,6))
plt.loglog(f_SI, abs(data), 'r')
plt.loglog(f_SI, psd**0.5, 'c')
plt.xlim(20,1e3)
plt.ylim(1e-24,1e-22)
plt.xlabel('$f$ [Hz]')
plt.ylabel('$h(f)$ and $S_h(f)$')

np.savetxt(out_file, np.c_[f_SI[:-1],datar[:-1],datai[:-1],psd[:-1]], header='f real_data imag_data psd')
```
